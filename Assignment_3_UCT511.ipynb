```python
a = int(input("Enter the number: "))
a_bin = bin(a)
print("Binary equivalent is: ",a_bin)
```

```python
a = int(input("Enter the number: "))
a_bin1 = ""
if a == 0:
  a_bin1 = "0"
while a > 0:
  rem = a % 2
  a_bin1 = str(rem) + a_bin1
  a //=2

print(f"Binary equivalent of {a} is {a_bin1}")
```

```python
while True:
  inp = input("Enter a mathematical expression (or quit to exit the calculator) : ")

  if inp.lower() == "quit":
    print("Calculator exiting!!!")
    break
  try:
    result = eval(inp)
    print("Result: ",result)
  except:
    print("Invalid Expression. Try Again!!!")
```

```python
print("(n(n-1))/2 =====> ((n*(n-1))/2) ")
n=int(input("Enter the value of n:"))
res = ((n*(n-1))/2)
print(f"Result of the mathematical expression for {n} is : {res}")
```

```python
pie=math.pi
print("4πr² =====> 4*π*r*r")
r=int(input("Enter the value of r: "))
res1 = 4 * pie * r * r
print(f"Result of the mathematical expression for r {r} is : {res1}")
```

```python
print("√(r(cos a)² + r(sin a)²)   =====> math.sqr(r*(math.cos(a))*(math.cos(a)) + r*(math.sin(a))*(math.sin(a)))")
r = int(input("Enter the value of r: "))
a = int(input("Enter the value of a: "))
a = math.radians(a)
res2 = math.sqrt(r*(math.cos(a))*(math.cos(a)) + r*(math.sin(a))*(math.sin(a)))
print(f"Result of the mathematical expression for r and a is : {res2}")
```

```python
print("(y2-y1)/(x2-x1)  =====>  (y2-y1)/(x2-x1)")
x1=int(input("Enter the value of x1"))
x2=int(input("Enter the value of x2"))
y1=int(input("Enter the value of y1"))
y2=int(input("Enter the value of y2"))
res3 =  (y2-y1)/(x2-x1)
print(f"Result of the mathematical expression for x1, x2, y1 and y2 is : {res3}")
```

```python
print("The sequence of numbers that would be generated by the range expression range(5):\n")
for i in range(5):
  print(i,"  ")

print("\nThe sequence of numbers that would be generated by the range expression range(3,10):\n")
for i in range(3,10):
  print(i,"  ")

print("\nThe sequence of numbers that would be generated by the range expression range(4,13,3):\n")
for i in range(4,13,3):
  print(i,"  ")

print("\nThe sequence of numbers that would be generated by the range expression range(15,5,-2):\n")
for i in range(15,5,-2):
  print(i,"  ")

print("\nThe sequence of numbers that would be generated by the range expression range(5,3):\n")
for i in range(5,3):
  print(i,"  ")
```
